```python
import matplotlib.pyplot as plt
import numpy as np

def my_function(x):
    return x ** 2

x_values = np.linspace(-10, 10, 100)

y_values = my_function(x_values)

plt.plot(x_values, y_values, label='y = x^2')
plt.title('Plot of my own function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
